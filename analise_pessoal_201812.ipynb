```python
recebeu_diaria = pessoal_df['Diárias'] > 0
```

```python
#Servidores que receberam maiores valores de Diária em Dez/2018
pessoal_df[recebeu_diaria].sort_values(by='Diárias', ascending=False)
```

```python
#Órgãos que mais pagaram diárias em Dez/2018
pessoal_df.groupby('Órgão/Entidade')['Diárias'].sum().sort_values(ascending=False)
```

```python
pessoal_df[recebeu_diaria].groupby('Órgão/Entidade')['Diárias'].agg([len, min, max, sum]).sort_values(by='sum', ascending=False)
```
