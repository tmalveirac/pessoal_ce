```python
recebeu_diaria = pessoal_df['Diárias'] > 0
pd.options.display.float_format = '${:,.2f}'.format
```

```python
#Servidores que mais receberam diárias em 2018
pessoal_df[recebeu_diaria].groupby(['Servidor','Órgão/Entidade', 'Cargo/Função/Emprego'])['Diárias'].sum().sort_values(ascending=False)
```

```python
#Servidores que receberam maiores valores de Diária em 2018 (Por mês)
pessoal_df[recebeu_diaria].sort_values(by='Diárias', ascending=False)
```

```python
#Órgãos que mais pagaram diárias em 2018
pessoal_df.groupby('Órgão/Entidade')['Diárias'].sum().sort_values(ascending=False)
```

```python
pessoal_df[recebeu_diaria].groupby('Órgão/Entidade')['Diárias'].agg([len, min, max, sum]).sort_values(by='sum', ascending=False)
```
